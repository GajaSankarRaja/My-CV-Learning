```python
import os
import numpy as np
import skimage
```

```python
input_dir='D:/clf-data'
category=['empty','not_empty']
```

```python
data,labels=[],[]
```

```python
from skimage.io import imread
from skimage.transform import resize
for ind,category in enumerate(category):
    for file in os.listdir(os.path.join(input_dir,category)):
        img_path=os.path.join(input_dir,category,file)
        img=imread(img_path)
        img=resize(img,(15,15))
        data.append(img.flatten())
        labels.append(ind)
data=np.array(data)
labels=np.array(labels)
```

```python
data
```

```python
labels
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(data,labels,test_size=0.25,random_state=42,shuffle=True,stratify=labels)
```

```python
from sklearn.svm import SVC
classifier=SVC()
classifier.fit(xtrain,ytrain)
```

```python
parameters=[{'gamma':[0.01,0.25,0.75,0.9],'C':[1,10,100,1000],'kernel':['linear','sigmoid','rbf']}]
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(classifier,parameters)
grid.fit(xtrain,ytrain)
```

```python
from sklearn.metrics import accuracy_score
best=grid.best_estimator_
ypred=best.predict(xtest)
accuracy_score(ytest,ypred)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(classifier,data,labels,cv=5)
```
