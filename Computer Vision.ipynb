```python
#Sobel Detector:
import numpy as np
import matplotlib.pyplot as plt
# Load the image
image_path = 'vijay.jpeg'  # Replace with your image path
image = cv.imread(image_path)
# Convert to grayscale
gray_image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
# Apply Gaussian smoothing (optional)
blurred_image = cv.GaussianBlur(gray_image, (3, 3), 0)
# Sobel operators
Gx = cv.Sobel(blurred_image, cv.CV_64F, 1, 0, ksize=3)
Gy = cv.Sobel(blurred_image, cv.CV_64F, 0, 1, ksize=3)
# Gradient magnitude
G = np.sqrt(Gx**2 + Gy**2)
# Normalize to range 0-255
Gx = np.uint8(255 * np.abs(Gx) / np.max(Gx))
Gy = np.uint8(255 * np.abs(Gy) / np.max(Gy))
G = np.uint8(255 * G / np.max(G))
# Display the results
plt.figure(figsize=(15, 10))
# Original image
plt.subplot(2, 2, 1)
plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')
# Gradient in X direction
plt.subplot(2, 2, 2)
plt.imshow(Gx, cmap='gray')
plt.title('Gradient in X direction')
plt.axis('off')
# Gradient in Y direction
plt.subplot(2, 2, 3)
plt.imshow(Gy, cmap='gray')
plt.title('Gradient in Y direction')
plt.axis('off')
# Edge-detected image
plt.subplot(2, 2, 4)
plt.imshow(G, cmap='gray')
plt.title('Sobel Edge Detection')
plt.axis('off')
plt.show()
```

```python
#Color Detector:
from PIL import Image
import cv2 as cv
import numpy as np
yellow=[0,255,255]
def get_limits(color):
    c=np.uint8([[color]])
    hsvC=cv.cvtColor(c,cv.COLOR_BGR2HSV)
    lower=hsvC[0][0][0]-10,100,100
    upper=hsvC[0][0][0]+10,255,255
    lower=np.array(lower,dtype=np.uint8)
    upper=np.array(upper,dtype=np.uint8)
    return lower,upper

cap=cv.VideoCapture(0)
while 1:
    ret,frame=cap.read()
    cv.imshow('image',frame)
    hsvImage=cv.cvtColor(frame,cv.COLOR_BGR2HSV)
    lower,upper=get_limits(color=yellow)
    mask=cv.inRange(hsvImage,lower,upper)
    mask_=Image.fromarray(mask)
    bbox=mask_.getbbox()
    #print(bbox)
    if bbox is not None:
        x1,y1,x2,y2=bbox
        frame=cv.rectangle(frame,(x1,y1),(x2,y2),(0,255,255),5)
    #frame=cv.resize(frame,(500,500))
    cv.imshow('image',frame)
    if cv.waitKey(1) &0xFF==ord('q'):
        break
cap.release()
cv.destroyAllWindows()
```

```python
import cv2 as cv
import numpy as np

yellow = [0, 255, 255]
def get_limits(color):
    c = np.uint8([[color]])
    hsvC = cv.cvtColor(c, cv.COLOR_BGR2HSV)
    lower = np.array([hsvC[0][0][0] - 10, 100, 100], dtype=np.uint8)
    upper = np.array([hsvC[0][0][0] + 10, 255, 255], dtype=np.uint8)
    return lower, upper
cap = cv.VideoCapture(0)
while True:
    ret, frame = cap.read()
    if not ret:
        break
    hsvImage = cv.cvtColor(frame, cv.COLOR_BGR2HSV)
    lower, upper = get_limits(color=yellow)
    mask = cv.inRange(hsvImage, lower, upper)
    contours, _ = cv.findContours(mask, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
    if contours:
        for cnt in contours:
            x, y, w, h = cv.boundingRect(cnt)
            cv.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 255), 5)

    cv.imshow('image', frame)
    
    if cv.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv.destroyAllWindows()
```
